```python
import pandas
```

```python
df = pandas.read_excel('남북한발전전력량.xls')
df
```

```python
new_df = df.iloc[[0,5],3:] # 합계만 추출 년도순으로
```

```python
new_df.columns # 컬럼명만 뽑아보기
```

```python
new_df.index = ['South','North'] # 인덱스만 뽑아서 수정
new_df
```

```python
# 컬럼을 정수형으로 바꿔주기 3가지 방법
new_df.columns = [int(x) for x in new_df.columns]  
# new_df.columns = new_df.columns.map(int)
# new_df.columns = new_df.columns.astype('int')
```

```python
new_df.transpose().plot() # 발전량 합계를 연도순으로 기본 선그래프
```

```python
new_df.transpose().plot(kind = 'bar') # 발전량 합계를 연도순으로 막대그래프
```
