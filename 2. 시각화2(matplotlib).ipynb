```python
import pandas
```

```python
df = pandas.read_excel('남북한발전전력량.xls')
df.head(10)
```

```python
df_ns = df.iloc[[0,5],2:] # 합계만 ([0,5]) 1990년도부터 (2:) 뽑아내기
df_ns.index = ['South', 'North']
df_ns
```

## 컬럼명 문자형에서 숫자로 바꾸기
1. [int(x) for x in df_ns.columns] <- 리스트 내포
2. list(map(int,df_ns.columns)) <- map()함수
3. df_ns.columns.astype(int) <- 판다스 astype()함수

```python
df_ns.columns = df_ns.columns.map(int)
df_ns.columns
```
